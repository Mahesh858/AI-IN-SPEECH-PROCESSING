```python
import numpy as np
from scipy.io import wavfile
import matplotlib.pyplot as plt
```

```python
sample_rate, signal = wavfile.read('/content/drive/MyDrive/mahesh.wav')
```

Find the first derivative of your speech signal with finite difference method. Listen to the first derivative signal and the original speech signal.

```python
time_step = 1 / sample_rate
first_derivative = np.diff(signal) / time_step
plt.subplot(2, 1, 1)
plt.plot(np.arange(len(signal)) / sample_rate, signal)
plt.title('Original Speech Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.subplot(2, 1, 2)
plt.plot(np.arange(len(first_derivative)) / sample_rate, first_derivative)
plt.title('First Derivative of Speech Signal')
plt.xlabel('Time (s)')
plt.ylabel('First Derivative')
plt.tight_layout()
plt.show()
wavfile.write('first_derivative.mp3', sample_rate, first_derivative.astype(np.int16))
```
